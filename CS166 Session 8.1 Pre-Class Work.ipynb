```
def calculate_q(k):
  '''
  Use a numerical root finder to determine q from the equation
  q = exp(k*(qÂ­1)).
  '''
  from scipy.optimize import root
  import numpy as np
  return root(lambda q: q - np.exp(k*(q-1)), 0).x[0]
```

```
vals = []
for i in range(1,11):
  vals.append(calculate_q(i))
```

```
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,9))
plot_x = np.linspace(0, 10, 10)
plt.xlabel('<k>')
plt.ylabel('q')
plt.plot(plot_x, vals)
plt.title('Graph of q for different values of <k> in the range [1, 10]')
```
